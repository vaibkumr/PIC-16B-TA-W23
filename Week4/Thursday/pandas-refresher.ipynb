```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
```

```python
data = load_iris()
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df['target'] = data['target']
```

```python
coefs = {}
for name, grouped_df in df.groupby(df['target']):
    X = grouped_df['sepal width (cm)'].values.reshape(-1, 1)
    y = grouped_df['petal length (cm)']
    reg = LinearRegression().fit(X, y)
    coefs[name] = reg.coef_[0]
```
