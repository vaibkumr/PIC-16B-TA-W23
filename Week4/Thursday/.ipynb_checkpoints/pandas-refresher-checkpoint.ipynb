```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
data = load_iris()
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df['target'] = data['target']
```

```python
for name, grouped_df in df.group_by('target'):
    print(name)
```
